```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from scipy.stats import skew

pd.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train = train.fillna(0)
test = test.fillna(0)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

train["form_field14"] = LabelEncoder().fit_transform(train["form_field14"])
train["default_status"] = LabelEncoder().fit_transform(train["default_status"])
test["form_field14"] = LabelEncoder().fit_transform(test["form_field14"])
```

```python
train.head()
```

```python
X = train.drop(["Applicant_ID", "default_status"], axis =1)
y = train["default_status"]

X_test = test.drop(["Applicant_ID"], axis = 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.45, stratify = y, random_state = 0)
## "stratify = y", to make sure the districution of the target "CHURN" are evenly distributed in the train and validation sets
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

prediction = model.predict(X_val)
score = roc_auc_score(y_val, prediction)
print("AUC Score on validation set: ",score)
```

```python
## we fit on the entire data this time
model.fit(X, y)
## we are now making prediction based on the "test dataset" stores with variable name "X_test" (if lost check the celss above)
## we are using "predict_proba()" can we are required to predict the likelihood(probabilitiy) of a user churning
## "[:,1]", this helps to select only the probabilites of churning (i.e 1), if we use "[:, 0]" it selects the probalities of not-churning (i.e 0)
pred = model.predict_proba(X_test)[:,1]
## so create a dictionary with the "user_id" from the "test_dataset" and the new predictions "pred"
## we trun the dictionary into a pandas dataframe using "pd.DataFrame()"
## we then export the dataframe into a csv file using ".to_csv"
pd.DataFrame({"Applicant_ID": test["Applicant_ID"], "default_status": pred}).to_csv("starter-submission_2.csv", index = False)
## gave a score of 0.7488 on the Public Leaderboard
```
